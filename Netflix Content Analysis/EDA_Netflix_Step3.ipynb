```python
# # ════════════════════════════════════════════════════════════
# # CHART 1 — Movies vs TV Shows Donut Chart
# # Libraries: numpy (value counts), matplotlib.pyplot (pie)
# # ════════════════════════════════════════════════════════════
# print("\nChart 1: Movies vs TV Shows ...")
 
# counts = df["type"].value_counts()
# labels = counts.index.tolist()
# values = counts.values                          # numpy array
 
# fig, ax = plt.subplots(figsize=(7, 7), facecolor="black")
 
# wedges, texts, autotexts = ax.pie(
#     values,
#     labels     = labels,
#     autopct    = "%1.1f%%",
#     startangle = 200,
#     colors     = [RED, BLUE],
#     wedgeprops = {"width": 0.55, "linewidth": 3},
#     textprops  = {"fontsize": 13},
# )
# for at in autotexts:
#     at.set_fontsize(15)
#     at.set_fontweight("bold")
#     at.set_color("white")
 
# ax.set_title("Movies vs TV Shows on Netflix", fontsize=16, fontweight="bold")
# ax.text(0, 0, f"{len(df):,}\nTitles",
#         ha="center", va="center", fontsize=14, fontweight="bold", color='white')
# ax.set_facecolor("white")
 
# plt.tight_layout()
# plt.savefig("chart1_movies_vs_tvshows.png")
# plt.close()
# print("  Saved chart1_movies_vs_tvshows.png")
import matplotlib.pyplot as plt

# --- Setup Colors & Data ---
BG_COLOR = "#1f1f1f"  # Dark grey background matching image_18f15f.png
RED = "#e50914"       # Netflix Red
BLUE = "#2b78b7"      # Blue from image_18f15f.png

# Ensure "Movie" is the primary slice for the center text logic
counts = df["type"].value_counts()
if "Movie" in counts.index and counts.index[0] != "Movie":
    counts = counts.reindex(["Movie", "TV Show"])

total_titles = len(df)
movie_pct = (counts.get("Movie", 0) / total_titles) * 100

# --- Plotting ---
# Increase height slightly to accommodate the bottom legend
fig, ax = plt.subplots(figsize=(7, 8), facecolor=BG_COLOR)
ax.set_facecolor(BG_COLOR)

# Create the donut chart
wedges, _ = ax.pie(
    counts.values,
    startangle=90,
    colors=[RED, BLUE],
    wedgeprops={"width": 0.4, "edgecolor": BG_COLOR, "linewidth": 1},
    counterclock=False
)

# Title - Top Left
plt.title("Content Type Split", color="white", loc="left", 
          fontsize=14, fontweight="bold", pad=20)

# Center Text
ax.text(0, 0.1, f"{movie_pct:.1f}%", 
        ha="center", va="center", color="white", 
        fontsize=18, fontweight="bold")
ax.text(0, -0.15, "Movies", 
        ha="center", va="center", color="grey", 
        fontsize=12)

# Manual Legend at Bottom Left
legend_x = -0.9
ax.text(legend_x, -1.2, f"●  Movies   {counts.get('Movie', 0):,}", color="grey", fontsize=11)
ax.text(legend_x, -1.35, f"●  TV Shows  {counts.get('TV Show', 0):,}", color="grey", fontsize=11)

# Overlay colored dots for the legend
ax.text(legend_x, -1.2, "●", color=RED, fontsize=11)
ax.text(legend_x, -1.35, "●", color=BLUE, fontsize=11)

plt.tight_layout()

# Save the file
plt.savefig("chart1_movies_vs_tvshows.png", facecolor=fig.get_facecolor(), bbox_inches='tight')

# Close the plot to prevent it from displaying in the UI
plt.close(fig)

print("Saved chart1_movies_vs_tvshows.png")
```

```python
# # CHART 2 — Content Growth Over Years (Stacked Bar)
# # Libraries: pandas (groupby/unstack), matplotlib.pyplot (bar)
# # ════════════════════════════════════════════════════════════
# print("Chart 2: Content growth over years ...")
 
# yearly = df.groupby(["year_added", "type"]).size().unstack(fill_value=0)
# yearly = yearly[yearly.index >= 2010]
 
# # numpy: get x positions for bars
# x      = np.arange(len(yearly.index))
# types  = yearly.columns.tolist()
# bottom = np.zeros(len(yearly))                  # numpy zeros for stacking
 
# fig, ax = plt.subplots(figsize=(12, 6), facecolor="white")
# colors = [RED, BLUE]
 
# for i, t in enumerate(types):
#     vals = yearly[t].values                     # numpy array
#     ax.bar(x, vals, bottom=bottom,
#            label=t, color=colors[i],
#            width=0.65, edgecolor="white", linewidth=0.5)
#     bottom += vals                              # numpy addition
 
# ax.set_xticks(x)
# ax.set_xticklabels([str(int(yr)) for yr in yearly.index], rotation=45, ha="right")
# ax.set_title("Netflix Content Added Per Year", fontsize=16, fontweight="bold")
# ax.set_xlabel("Year Added", fontsize=12)
# ax.set_ylabel("Number of Titles", fontsize=12)
# ax.legend(title="Content Type", frameon=False, fontsize=11)
# ax.set_facecolor("white")
 
# plt.tight_layout()
# plt.savefig("chart2_content_growth.png", dpi=150, bbox_inches="tight")
# plt.close()
# print("  Saved chart2_content_growth.png")
print("Chart 2: Content growth over years ...")

yearly = df.groupby(["year_added", "type"]).size().unstack(fill_value=0)
yearly = yearly[yearly.index >= 2015]   # match visible years

x = np.arange(len(yearly.index))
types = yearly.columns.tolist()
bottom = np.zeros(len(yearly))

# 🔴 Dark theme figure
fig, ax = plt.subplots(figsize=(12, 6), facecolor="#1c1c1c")
ax.set_facecolor("#1c1c1c")

colors = ["#E50914", "#4C9BE8"]   # Netflix red + blue

# 🔹 Plot stacked bars
for i, t in enumerate(types):
    vals = yearly[t].values
    bars = ax.bar(
        x, vals, bottom=bottom,
        label=t,
        color=colors[i],
        width=0.65
    )
    bottom += vals

# 🔹 Add total labels on top of bars
totals = yearly.sum(axis=1).values
for i, total in enumerate(totals):
    ax.text(
        x[i], total + 50, f"{total:,}",
        ha="center", va="bottom",
        fontsize=9, color="white", fontweight="bold"
    )

# 🔹 X-axis formatting
ax.set_xticks(x)
ax.set_xticklabels(yearly.index.astype(int), color="white")

# 🔹 Titles & labels
ax.set_title("Titles Added Per Year  (Stacked by Type)",
             fontsize=14, color="white", pad=15)

# 🔹 Remove y-axis (to match style)
ax.yaxis.set_visible(False)

# 🔹 Remove spines
for spine in ax.spines.values():
    spine.set_visible(False)

# 🔹 Legend (top-right, horizontal)
ax.legend(
    loc="upper right",
    frameon=False,
    fontsize=9,
    labelcolor="white"
)

plt.tight_layout()
plt.savefig("chart2_content_growth.png", dpi=150, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.close()

print("  Saved chart2_content_growth.png")
```

```python
# # CHART 4 — Monthly Additions Heatmap (Month x Year)
# # Libraries: pandas (pivot_table), seaborn (heatmap)
# # ════════════════════════════════════════════════════════════
# print("Chart 4: Monthly heatmap ...")
 
# heat = df[df["year_added"] >= 2015].copy()
# pivot = heat.pivot_table(
#     index    = "month_name",
#     columns  = "year_added",
#     values   = "show_id",
#     aggfunc  = "count",
#     fill_value = 0
# )
 
# month_order = ["January","February","March","April","May","June",
#                "July","August","September","October","November","December"]
# pivot = pivot.reindex([m for m in month_order if m in pivot.index])
 
# # numpy: confirm data shape for annotation
# print(f"  Heatmap shape: {pivot.shape}")
 
# fig, ax = plt.subplots(figsize=(12, 7), facecolor="white")
 
# sns.heatmap(
#     pivot,
#     ax         = ax,
#     cmap       = "YlOrRd",
#     annot      = True,
#     fmt        = "d",
#     annot_kws  = {"size": 9},
#     linewidths = 0.5,
#     linecolor  = "white",
#     cbar_kws   = {"label": "Titles Added"},
# )
# ax.set_title("Netflix Content Additions by Month & Year",
#              fontsize=16, fontweight="bold")
# ax.set_xlabel("Year Added", fontsize=12)
# ax.set_ylabel("Month", fontsize=12)
 
# plt.tight_layout()
# plt.savefig("chart4_monthly_heatmap.png", dpi=150, bbox_inches="tight")
# plt.close()
# print("  Saved chart4_monthly_heatmap.png")
# CHART 4 — Monthly Additions Heatmap (Styled like reference)
print("Chart 4: Monthly heatmap ...")

heat = df[df["year_added"] >= 2018].copy()

pivot = heat.pivot_table(
    index="month_name",
    columns="year_added",
    values="show_id",
    aggfunc="count",
    fill_value=0
)

# Order months
month_order = ["January","February","March","April","May","June",
               "July","August","September","October","November","December"]
pivot = pivot.reindex([m for m in month_order if m in pivot.index])

print(f"  Heatmap shape: {pivot.shape}")

# Short month labels (Jan, Feb, ...)
pivot.index = [m[:3] for m in pivot.index]

# Figure setup (dark theme)
fig, ax = plt.subplots(figsize=(8, 6), facecolor="#1e1e1e")
ax.set_facecolor("#1e1e1e")

# Heatmap
sns.heatmap(
    pivot,
    ax=ax,
    cmap="Reds",              # strong red gradient like image
    annot=True,
    fmt="d",
    annot_kws={"size": 10, "color": "white", "weight": "bold"},
    linewidths=2,             # thicker gaps → box-like feel
    linecolor="#1e1e1e",      # blend with background
    cbar=False                # remove color bar (not in image)
)

# Titles & labels
ax.set_title("Monthly Additions Heatmap",
             fontsize=14, color="white", pad=12, weight="bold")

ax.set_xlabel("")
ax.set_ylabel("")

# Tick styling
ax.tick_params(axis='x', colors='lightgray', labelsize=11)
ax.tick_params(axis='y', colors='lightgray', labelsize=11, rotation=0)

# Move x-axis labels to top (like image)
ax.xaxis.tick_top()
ax.xaxis.set_label_position('top')

# Remove spines
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.savefig("chart4_monthly_heatmap.png", dpi=150,
            bbox_inches="tight", facecolor=fig.get_facecolor())
plt.close()

print("  Saved chart4_monthly_heatmap.png")
```

```python
# CHART 6 — Rating Distribution (Bar Chart)
# Libraries: pandas (value_counts), seaborn (barplot)
# ════════════════════════════════════════════════════════════
print("Chart 6: Rating distribution ...")
 
ratings        = df["rating"].value_counts().reset_index()
ratings.columns = ["rating", "count"]
 
# numpy: build palette array
palette6 = [RED if i == 0 else "#888888" for i in range(len(ratings))]
 
fig, ax = plt.subplots(figsize=(10, 5), facecolor="white")
 
sns.barplot(
    data    = ratings,
    x       = "rating",
    y       = "count",
    palette = palette6,
    ax      = ax,
    order   = ratings["rating"].tolist(),
)
ax.set_title("Content Rating Distribution on Netflix",
             fontsize=16, fontweight="bold")
ax.set_xlabel("Rating", fontsize=12)
ax.set_ylabel("Number of Titles", fontsize=12)
 
for i, row in ratings.iterrows():
    ax.text(i, row["count"] + 20, f'{row["count"]:,}',
            ha="center", fontsize=9)
 
ax.set_facecolor(LIGHT_GRAY)
 
plt.tight_layout()
plt.savefig("chart6_rating_dist.png", dpi=150, bbox_inches="tight")
plt.close()
print("  Saved chart6_rating_dist.png")
 
```

```python
# # CHART 7 — Movie Duration Distribution (Histogram)
# # Libraries: numpy (mean/median/filter), seaborn (histplot),
# #            matplotlib.pyplot (axvline, legend)
# # ════════════════════════════════════════════════════════════
# print("Chart 7: Movie duration distribution ...")
 
# dur_raw  = df[df["type"] == "Movie"]["duration_minutes"].dropna()
 
# # numpy: filter outliers using boolean mask
# dur_arr  = dur_raw.values                          # convert to numpy array
# mask     = (dur_arr >= 30) & (dur_arr <= 240)
# dur_filt = dur_arr[mask]
 
# mean_dur   = np.mean(dur_filt)
# median_dur = np.median(dur_filt)
 
# fig, ax = plt.subplots(figsize=(10, 5), facecolor="white")
 
# sns.histplot(dur_filt, bins=40, color=RED,
#              edgecolor="white", linewidth=0.5, ax=ax)
 
# ax.axvline(mean_dur,   color=BLUE,   linestyle="--", linewidth=2,
#            label=f"Mean:   {mean_dur:.0f} min")
# ax.axvline(median_dur, color=ORANGE, linestyle="--", linewidth=2,
#            label=f"Median: {median_dur:.0f} min")
 
# ax.set_title("Movie Duration Distribution", fontsize=16, fontweight="bold")
# ax.set_xlabel("Duration (minutes)", fontsize=12)
# ax.set_ylabel("Number of Movies", fontsize=12)
# ax.legend(fontsize=11, frameon=False)
# ax.set_facecolor(LIGHT_GRAY)
 
# plt.tight_layout()
# plt.savefig("chart7_movie_duration.png", dpi=150, bbox_inches="tight")
# plt.close()
# print("  Saved chart7_movie_duration.png")
# CHART 7 — Movie Duration Distribution (Improved Styling)
print("Chart 7: Movie duration distribution ...")

# Filter movie durations
dur_raw = df[df["type"] == "Movie"]["duration_minutes"].dropna()

# Convert to numpy and filter outliers
dur_arr = dur_raw.values
dur_filt = dur_arr[(dur_arr >= 30) & (dur_arr <= 240)]

# Stats
mean_dur = np.mean(dur_filt)
median_dur = np.median(dur_filt)

# Figure setup (dark theme like your chart)
fig, ax = plt.subplots(figsize=(10, 5), facecolor="#1e1e1e")
ax.set_facecolor("#1e1e1e")

# Histogram
sns.histplot(
    dur_filt,
    bins=40,
    color="red",
    edgecolor=None,
    alpha=0.9,
    ax=ax
)

# Mean & Median lines
ax.axvline(mean_dur, color="cyan", linestyle="-", linewidth=2,
           label=f"Mean: {mean_dur:.0f} min")

ax.axvline(median_dur, color="orange", linestyle="-", linewidth=2,
           label=f"Median: {median_dur:.0f} min")

# Titles & labels
ax.set_title("Movie Duration Distribution", fontsize=14, color="white", pad=10)
ax.set_xlabel("Duration (minutes)", color="lightgray")
ax.set_ylabel("Number of Movies", color="lightgray")

# Tick styling
ax.tick_params(colors="lightgray")

# Legend styling
legend = ax.legend(frameon=False, fontsize=10)
for text in legend.get_texts():
    text.set_color("white")

# Remove spines for cleaner look
for spine in ax.spines.values():
    spine.set_visible(False)

plt.tight_layout()
plt.savefig("chart7_movie_duration.png", dpi=150, bbox_inches="tight", facecolor=fig.get_facecolor())
plt.close()

print("  Saved chart7_movie_duration.png")
```

```python
# # CHART 8 — Top 5 Genre Trends Over Years (Line Chart)
# # Libraries: pandas (groupby/filter), numpy (arange),
# #            matplotlib.pyplot (plot/xticks)
# # ════════════════════════════════════════════════════════════
# print("Chart 8: Top 5 genre trends ...")
 
# top5      = g["genre"].value_counts().head(5).index.tolist()
# g_trend   = g[g["genre"].isin(top5)].copy()
# g_grouped = (g_trend.groupby(["year_added", "genre"])
#                     .size()
#                     .reset_index(name="count"))
# g_grouped = g_grouped[g_grouped["year_added"] >= 2015]
 
# palette8  = [RED, BLUE, ORANGE, GREEN, PURPLE]
 
# fig, ax = plt.subplots(figsize=(12, 6), facecolor="white")
 
# for i, genre in enumerate(top5):
#     subset = g_grouped[g_grouped["genre"] == genre]
#     years  = subset["year_added"].values           # numpy array
#     cnts   = subset["count"].values                # numpy array
#     ax.plot(years, cnts,
#             marker="o", linewidth=2.5,
#             color=palette8[i], label=genre, markersize=7)
 
# # numpy: build integer x-ticks
# all_years = np.sort(g_grouped["year_added"].unique())
# ax.set_xticks(all_years)
# ax.set_xticklabels([str(int(y)) for y in all_years])
 
# ax.set_title("Top 5 Genre Trends (2015–2021)", fontsize=16, fontweight="bold")
# ax.set_xlabel("Year Added", fontsize=12)
# ax.set_ylabel("Number of Titles", fontsize=12)
# ax.legend(frameon=False, fontsize=10, loc="upper left")
# ax.set_facecolor(LIGHT_GRAY)
 
# plt.tight_layout()
# plt.savefig("chart8_genre_trends.png", dpi=150, bbox_inches="tight")
# plt.close()
# print("  Saved chart8_genre_trends.png")
print("Chart 8: Top 5 genre trends ...")

top5      = g["genre"].value_counts().head(5).index.tolist()
g_trend   = g[g["genre"].isin(top5)].copy()

g_grouped = (g_trend.groupby(["year_added", "genre"])
                    .size()
                    .reset_index(name="count"))

g_grouped = g_grouped[g_grouped["year_added"] >= 2015]

# 🎨 Bright colors (match visual)
palette8 = ["#FF2E2E", "#4C9BE8", "#F5A623", "#2ECC71", "#9B59B6"]

# 🌑 Dark theme
fig, ax = plt.subplots(figsize=(12, 6), facecolor="#1c1c1c")
ax.set_facecolor("#1c1c1c")

# 🔹 Plot lines
for i, genre in enumerate(top5):
    subset = g_grouped[g_grouped["genre"] == genre]
    years  = subset["year_added"].values
    cnts   = subset["count"].values

    ax.plot(
        years, cnts,
        marker="o",
        linewidth=2.5,
        markersize=6,
        color=palette8[i],
        label=genre
    )

# 🔹 X-axis ticks
all_years = np.sort(g_grouped["year_added"].unique())
ax.set_xticks(all_years)
ax.set_xticklabels(all_years.astype(int), color="white")

# 🔹 Remove Y-axis labels (clean look like image)
ax.yaxis.set_visible(False)

# 🔹 Subtle grid (horizontal only)
ax.grid(axis="y", linestyle="--", alpha=0.2)

# 🔹 Remove spines (borders)
for spine in ax.spines.values():
    spine.set_visible(False)

# 🔹 Title (left aligned)
ax.set_title("Top 5 Genre Trends (2015–2021)",
             fontsize=13,
             color="white",
             loc="left",
             pad=15)

# 🔹 Legend styling
legend = ax.legend(
    frameon=False,
    fontsize=9,
    loc="upper left"
)

# make legend text white
for text in legend.get_texts():
    text.set_color("white")

plt.tight_layout()
plt.savefig("chart8_genre_trends.png",
            dpi=150,
            bbox_inches="tight",
            facecolor=fig.get_facecolor())
plt.close()

print("  Saved chart8_genre_trends.png") 
```
